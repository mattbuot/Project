```python
import pandas as pd
import re
import requests
```

```python
pd.set_option('display.max_columns', 173)
```

```python
df = pd.read_csv('data/en.openfoodfacts.org.products.csv', sep='\t', low_memory=False, encoding='utf-8')
```

```python
df.head()
```

```python
def without_nan(df, column):
    return df[~df[column].isnull()]
```

```python
# Selecting labels with an organic label
def organic_labels(df):
    return df['labels'][
        df['labels'].str.contains("Bio", na = False, flags = re.IGNORECASE)
        | df['labels'].str.contains("Organic", na = False, flags = re.IGNORECASE)
        |  df['labels'].str.contains("AB", na = False, flags = re.IGNORECASE)
        ].unique()

# Selecting products with an organic label
def organic_products(df):
    return df[
        df['labels'].str.contains("Bio", na = False, flags = re.IGNORECASE)
        | df['labels'].str.contains("Organic", na = False, flags = re.IGNORECASE)
        |  df['labels'].str.contains("AB", na = False, flags = re.IGNORECASE)
        ]
```

```python
print('There are', len(organic_products(df)), 'organic products in the database')
```

```python
# Retrieving front image url from barcode
def get_image_url(code):
    url = 'https://world.openfoodfacts.org/api/v0/product/' + str(code) + '.json'
    r = requests.get(url)
    json = r.json()
    if r.status_code == 200 and json['status_verbose'] == 'product found':
        try:
            return json['product']['image_url']
        except KeyError:
            try:
                return json['product']['selected_images']['front']['display']['fr']
            except KeyError:
                return 'No image available'
    else:
        return 'No image available'
```

```python
# Saving image locally from url and 
def save_image(name, image_url):
    image_data = requests.get(image_url).content
    images_folder = 'images/'
    file_name = name + '.jpg'
    with open(images_folder + file_name, 'wb') as handler:
        handler.write(image_data)
```

```python
# Downloading all images corresponding to organic products in an 'images' folder which needs to be created beforehand
def download_organic_images(n=0):
    i = 0

    if n > 0:
        barcodes = organic_products(df).head(n)['code']
    else:
        barcodes = organic_products(df)['code']
    
    length = len(barcodes)

    for code in barcodes:
        url = get_image_url(code)
    
        if url != 'No image available':
            save_image(code, url)
            i = i + 1
    
        progress = str(i) + '/' + str(length)
        print(progress, end='\r')
    
    print('Download completed!')
```

```python
download_organic_images(10)
```
